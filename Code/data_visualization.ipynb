```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from collections import Counter
```

```python
data = pd.read_csv("bbc_data.csv")
```

```python
data.head()
```

```python
class_counts = data['labels'].value_counts()
colors = plt.cm.Set3(range(len(class_counts)))
plt.figure(figsize=(8, 6))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Class Distribution')
plt.show()
```

```python
words = ' '.join(data['data'])
wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='magma').generate(words)
plt.figure(figsize=(12, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('Word Distribution')
plt.axis('off')
plt.show()
```
