```python
import torch
from transformers import BertForSequenceClassification
from transformers import BertTokenizer
from sklearn.preprocessing import LabelEncoder
```

```python
model_path = "/trained_model"
model = BertForSequenceClassification.from_pretrained(model_path)
```

```python
labels_text = ["entertainment", "sport", "business", "politics", "tech"]
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(labels_text)
print("Encoded labels:", encoded_labels)
```

```python
encoded_labels = [1, 3, 0, 2, 4]
labels_text = ["entertainment", "sport", "business", "politics", "tech"]

label_decoder = {}
for i in range(len(encoded_labels)):
    label_decoder[encoded_labels[i]] = labels_text[i]

print("Encoded labels:", encoded_labels)
print("Decoded labels:")
for label in encoded_labels:
    print(label, "->", label_decoder[label])
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
def predict_text(model, text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
    outputs = model(**inputs)
    _, preds = torch.max(outputs.logits, 1)
    return preds.item()
```

```python
text = "The latest smartphone model features advanced facial recognition technology."
predicted_class = predict_text(model, text)
print("Predicted Class:", label_decoder[predicted_class])
```

```python
text = "The new movie release received mixed reviews from critics."
predicted_class = predict_text(model, text)
print("Predicted Class:", label_decoder[predicted_class])
```
